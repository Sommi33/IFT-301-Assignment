```python
import matplotlib.pyplot as plt

# Data for the bar chart
categories= ['Mushroom', 'Pineapple', 'Prawns', 'Sausage', 'Spinach']
proportions = [0.16, 0.28, 0.08, 0.18, 0.28]

# Bar chart
plt.barh(categories, proportions, color='red', edgecolor='black')
plt.xlabel('Proportion of Total')
plt.ylabel('Categories')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Data from the table
data = {
    'Gender': ['Female']*10 + ['Male']*10,
    'Height': [175, 158, 166, 175, 160, 165, 166, 170, 170, 172,
               184, 182, 180, 191, 189, 181, 180, 170, 176, 185],
    'Breath_Held': [22.22, 30.57, 17.47, 22.39, 26.90, 36.85, 27.33, 29.55, 13.87, 34.66,
                    60.75, 67.41, 42.19, 59.74, 52.64, 43.37, 73.27, 59.09, 51.15, 58.32]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Plot 1: Histogram of breath-holding times
plt.figure(figsize=(8, 5))
plt.hist(df['Breath_Held'], bins=8, color='red', edgecolor='black')
plt.title('Histogram of Breath-Holding Times')
plt.xlabel('Breath-Holding Time (s)')
plt.ylabel('Frequency')
plt.show()

# Plot 2: Side-by-side dot plots for males and females
plt.figure(figsize=(8, 5))
male_times = df[df['Gender'] == 'Male']['Breath_Held']
female_times = df[df['Gender'] == 'Female']['Breath_Held']

# Add female times
plt.scatter(female_times, ['Female']*len(female_times), color='pink', label='Female', alpha=0.7)
# Add male times
plt.scatter(male_times, ['Male']*len(male_times), color='skyblue', label='Male', alpha=0.7)

plt.title('Side-by-Side Dot Plot of Breath-Holding Times')
plt.xlabel('Breath-Holding Time (s)')
plt.ylabel('Gender')
plt.legend()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data from the table
music_growth = [304, 257, 174, 214, 69, 317, 387, 47, 157, 0, 352, 308, 317, 286, 236, 299, 206, 278, 188, 43, 0, 0, 0, 0, 0]
no_music_growth = [292, 270, 47, 288, 324, 292, 364, 316, 287, 75, 282, 149, 274, 319, 213, 3, 524, 2, 128, 219, 94, 164, 0, 0, 0]

# Dot Plot
plt.figure(figsize=(12, 6))
plt.plot(music_growth, [0] * len(music_growth), 'o', label='With Music', color='green', alpha=0.7)
plt.plot(no_music_growth, [1] * len(no_music_growth), 'o', label='Without Music', color='red', alpha=0.7)

# Customize plot
plt.title('Dot Plot of Plant Growth with and without Music')
plt.xlabel('Plant Growth (mm)')
plt.yticks([0, 1], ['With Music', 'Without Music'])
plt.legend()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

# Histogram
plt.figure(figsize=(12, 6))
plt.hist(music_growth, bins=10, alpha=0.7, label='With Music', color='green', edgecolor='black')
plt.hist(no_music_growth, bins=10, alpha=0.7, label='Without Music', color='red', edgecolor='black')

# Customize plot
plt.title('Histogram of Plant Growth with and without Music')
plt.xlabel('Plant Growth (mm)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
